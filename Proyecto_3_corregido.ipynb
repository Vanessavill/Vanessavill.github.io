```python
class objeto:
  def __init__ (self, t):
    self.t=t

  def registros(self):
    try:
      reg=int(input('¿A qué registro quiere acceder? Ingrese un número: '))
      return self.t[reg]
    except ValueError as e:
      print('Hay un error', e)
    except NameError:
      print('La variable no esta definida')

  def columnas(self, nombre_columna, datos):
      datos_columna=[]
      pos=datos[0].index(nombre_columna)
      for registro in datos[1:]:
        self.columnas=datos_columna.append(registro[pos])
      return datos_columna

  def filtro(self, nombre_columna, condicional):
    v=[]
    fil= self.columnas(nombre_columna)
    fil = [int(i) for i in fil]
    v.append(list(filter(lambda t: eval(condicional), fil)))
    return v

  def mean(self):
    return round(sum(self.t)/len(self.t))

  def var(self):
    media=sum(self.t)/len(self.t)
    return sum((i-media)**2 for i in self.t)/(len(self.t)-1)

  def std(self):
    return (self.var()**(0.5))

  def coef_var (self):
    return self.std()/self.mean()

  def median(self):
    half = len(self.t)//2
    if not len(self.t) % 2:
      return (self.t[half - 1]+self.t[half]) / 2.0
    return self.t[half]
```
