```python
df_1=df.drop(['ESTU_AREARESIDE', 'ESTU_CODDANE_COLE_TERMINO', 'ESTU_COD_COLE_MCPIO_TERMINO', 
              'ESTU_TIPODOCUMENTOSB11', 'FAMI_TRABAJOLABORPADRE', 'FAMI_TRABAJOLABORMADRE', 'ESTU_OTROCOLE_TERMINO', 'ESTU_HORASSEMANATRABAJA', 
              'FAMI_TIENELAVADORA', 'FAMI_TIENEHORNOMICROOGAS','FAMI_TIENESERVICIOTV', 'FAMI_TIENEAUTOMOVIL', 'FAMI_TIENEMOTOCICLETA',
              'FAMI_TIENECONSOLAVIDEOJUEGOS', 'FAMI_CUANTOSCOMPARTEBAÑO',
              'ESTU_PRESENTACIONCASA', 'ESTU_AGREGADO','ESTU_TITULOOBTENIDOBACHILLER', 'ESTU_COLE_TERMINO'], axis=1)
```

```python
cat = df_1.select_dtypes('object').columns
for col in cat:
  df_1[cat] = df_1[cat].astype('category') #Cambiamos todos los tipos de datos 'Object' a 'Category' para poder tratar los datos nulos.
```

```python
df_1['ESTU_VALORMATRICULAUNIVERSIDAD'].value_counts(dropna=False, normalize=True)
```

```python
df_1['ESTU_VALORMATRICULAUNIVERSIDAD'].fillna(df_1['ESTU_VALORMATRICULAUNIVERSIDAD'].mode()[0], inplace=True)
```

```python
df_1['ESTU_PAGOMATRICULABECA'].value_counts(dropna=False, normalize=True)
```

```python
df_1['ESTU_PAGOMATRICULABECA'].fillna(df_1['ESTU_PAGOMATRICULABECA'].mode()[0], inplace=True)
```

```python
df_1['ESTU_PAGOMATRICULACREDITO'].fillna(df_1['ESTU_PAGOMATRICULACREDITO'].mode()[0], inplace=True)
df_1['ESTU_PAGOMATRICULAPADRES'].fillna(df_1['ESTU_PAGOMATRICULAPADRES'].mode()[0], inplace=True)
df_1['ESTU_PAGOMATRICULAPROPIO'].fillna(df_1['ESTU_PAGOMATRICULAPROPIO'].mode()[0], inplace=True)
df_1['ESTU_COMOCAPACITOEXAMENSB11'].fillna(df_1['ESTU_COMOCAPACITOEXAMENSB11'].mode()[0], inplace=True)
df_1['ESTU_SEMESTRECURSA'].fillna(df_1['ESTU_SEMESTRECURSA'].mode()[0], inplace=True)
df_1['FAMI_EDUCACIONPADRE'].fillna(df_1['FAMI_EDUCACIONPADRE'].mode()[0], inplace=True)
df_1['FAMI_EDUCACIONMADRE'].fillna(df_1['FAMI_EDUCACIONMADRE'].mode()[0], inplace=True)
df_1['FAMI_ESTRATOVIVIENDA'].fillna(df_1['FAMI_ESTRATOVIVIENDA'].mode()[0], inplace=True)
df_1['FAMI_TIENEINTERNET'].fillna(df_1['FAMI_TIENEINTERNET'].mode()[0], inplace=True)
df_1['FAMI_TIENECOMPUTADOR'].fillna(df_1['FAMI_TIENECOMPUTADOR'].mode()[0], inplace=True)
df_1['ESTU_PAISDOCUMENTOS11'].fillna(df_1['ESTU_PAISDOCUMENTOS11'].mode()[0], inplace=True)
df_1['MOD_COMUNI_ESCRITA_DESEM'].fillna(df_1['MOD_COMUNI_ESCRITA_DESEM'].mode()[0], inplace=True)
df_1['ESTU_INSE_INDIVIDUAL'].fillna(df_1['ESTU_INSE_INDIVIDUAL'].mode()[0], inplace=True)
df_1['ESTU_NSE_INDIVIDUAL'].fillna(df_1['ESTU_NSE_INDIVIDUAL'].mode()[0], inplace=True)
```

```python
for col, por in ((df_1.isna().sum() / df_1.shape[0]) * 100).to_dict().items():
  print(f"En la columna '{col}' hace falta el {round(por,4)}% de los datos") #Verificar que los Missing Values sean 0 
```
