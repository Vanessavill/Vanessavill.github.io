```python
from scipy.spatial.distance import mahalanobis
from scipy.stats import chi2
import scipy.stats as stats
```

```python
df_mahalanobis=df[['price','minimum_nights', 'number_of_reviews','reviews_per_month', 'calculated_host_listings_count',
       'availability_365']] #Se tienen en cuenta únicamente las variables númericas
```

```python
mahal_distances = []
for row in df_mahalanobis.to_numpy():
  mahal_distances.append(mahalanobis(row, df_mahalanobis.mean(), np.linalg.inv(np.cov(df_mahalanobis.values.T))))
```

```python
df['mahal_distances'] = mahal_distances
df['p_value'] = 1 - stats.chi2.cdf(df['mahal_distances'], 13-1)
df.head()
```
